```python
# Download neccessary data
import pandas as pd

# We specify the exact filename we have used.
file_path = 'Phishing_Legitimate_full.csv'
df = pd.read_csv(file_path)

# This checks the dimensions (rows, columns) of the raw data.
print(f"Dataset loaded with {df.shape[0]} samples and {df.shape[1]} features.")

# Cleaning the data
# Identifying and handling missing values (NaNs).
null_count = df.isnull().sum().sum()
if null_count > 0:
    print(f"Detected {null_count} null values. Applying mean imputation.")
    df = df.fillna(df.mean())
else:
    print("Integrity Check: No missing values detected.")

# removing the non-statistical columns
# The 'id' column is a unique identifier; it does not contain patterns
# related to phishing and would lead to 'overfitting' if left in.
if 'id' in df.columns:
    df = df.drop(columns=['id'])
    print("Column 'id' dropped to focus on predictive features.")

# Removing Deduplication.
# Repeating rows can bias the model toward specific samples.
before_dup = len(df)
df = df.drop_duplicates()
print(f"Removed {before_dup - len(df)} duplicate rows.")

# Seperating everthing else from the target
# We isolate the 'Target' (CLASS_LABEL) from the 'Features' (everything else).
X = df.drop(columns=['CLASS_LABEL'])
y = df['CLASS_LABEL']

print("Data cleaning complete. Ready for model training.")
```
